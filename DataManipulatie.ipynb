```python
from geopy.extra.rate_limiter import RateLimiter
from geopy import Nominatim
locator = Nominatim(user_agent="myGeocoder", timeout=1)

import fiona

# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode)

def getLatLong(z):
    # 2- - create location column
    z['location'] = z['Hometown'].apply(geocode)
    # 3 - create longitude, laatitude and altitude from location column (returns tuple)
    z['point'] = z['location'].apply(lambda loc: tuple(loc.point) if loc else None)
    # 4 - split point column into latitude, longitude and altitude columns
    z[['latitude', 'longitude', 'altitude']] = pd.DataFrame(z['point'].tolist(), index=z.index)
    z = z.drop(['point', 'location', 'Body'], axis=1)
    z = z[z['latitude'].notna()]
    z['Country'] = z.apply(lambda row: getCountry(row), axis = 1)
    return z
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="myGeocoder", timeout=1)
def getCountry(row):
    lat = float(row.latitude)
    long = float(row.longitude)

    location = geolocator.reverse([lat, long])
    country = location.raw['address']['country']
    return country
```
