```python
SurvivalList = [DeckCrew,
        EngineeringCrew,
        Officers,
        RestaurantCrew,
        VictuallingCrew,
        FirstClass,
        SecondClass,
        ThirdClass]

DeadList  = [OrchestraCrew,
             PostalCrew,
             RestaurantCrew]

DataList = [DeckCrew,
            EngineeringCrew,
            Officers,
            RestaurantCrew,
            VictuallingCrew,
            FirstClass,
            SecondClass,
            ThirdClass,
            OrchestraCrew,
            PostalCrew,
            RestaurantCrew]
```

```python
def addSurvival(data):
    data['Survived'] = pd.notnull(data["Lifeboat"])
    print(data)
    return data

def deadSurvival(data):
    data['Survived'] = False
    print(data)
    return data
```

```python
for dataframe in SurvivalList:
    dataframe = addSurvival(dataframe)

for dataframe in DeadList:
    dataframe = deadSurvival(dataframe)
```
